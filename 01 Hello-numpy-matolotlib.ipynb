```python
import numpy as np
data=np.array([4, 5, 6, 7, 8])
print('original_data:',data)
print('squared', data**2)
print('mean:', np.mean(data))
```

```python
import matplotlib.pyplot as plt
plt.plot(data, data** 2,marker='o',color='purple')
plt.title('my first plot')
plt.xlabel('values')
plt.ylabel('square')
plt.grid(True)
plt.show()
```
